```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
```

### Below is a dry test to understand how a model is used, it is not specific(random model) used for classification purpose.So at the end it gives 0 and 1 as output(binary prediction).


```python
model_name = "bert-base-uncased"
model = AutoModelForSequenceClassification.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
text = """
Artificial intelligence is transforming the way humans interact with technology. 
From natural language processing to computer vision, modern machine learning models 
can understand text, generate realistic images, and even assist in medical diagnosis. 
As these systems become more advanced, it is important to ensure they are used 
responsibly, ethically, and in ways that benefit society as a whole.
"""
```

```python
inputs = tokenizer(text, return_tensors="pt")
```

```python
outputs = model(**inputs)
```

```python
outputs
```

```python
outputs.logits
```

```python
import torch
predictions = torch.argmax(outputs.logits)
print(f"Predicted class: {predictions.item()}")
```

### below is an example of sentimental analysis

```python
from transformers import pipeline

sent = "distilbert-base-uncased-finetuned-sst-2-english"
classifier = pipeline("sentiment-analysis", model=sent)
```

```python
texts = [
    "I love making models!",
    "The weather today is terrible.",
    "I fucking HATE you."
]
```

```python
results = classifier(texts)
```

```python
for t,result in zip(texts,results):
    print(f"Text: {t}")
    print(f"Label: {result['label']}, Score: {result['score']:.3f}\n")
```
