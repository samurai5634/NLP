```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
data = {
    "feature_1" : [1,2,3,4,5],
    "feature_2" : [5,4,2,3,1],
    "label" : [0,0,1,1,0]
}

```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
x = df[["feature_1","feature_2"]]
y = df["label"]
```

```python
x
```

```python
mod = Sequential()
mod.add(Dense(10,input_dim =2,activation="relu"))
mod.add(Dense(1,activation="sigmoid"))
```

```python
mod.compiled_loss
```

```python
mod.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
mod.fit(x,y,verbose = 2, epochs= 100, batch_size= 1)
```

```python
test_data = np.array([[0.2, 0.4]])
prediction = mod.predict(test_data)
predicted_label = (prediction > 0.5).astype(int)
```
