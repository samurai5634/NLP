### Named Entity Recognition

```python
import pandas as pd 
import spacy 
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
content  = "Last Friday, Elon Musk visited Berlin, Germany to attend the launch event of Tesla's new Model Z. He met with Angela Merkel at the Brandenburg Gate and discussed potential investments in renewable energy. Meanwhile, Apple Inc. announced a collaboration with BMW to develop new electric vehicle technology. The event was covered by Reuters and The New York Times."
doc = nlp(content)
```

```python
for e in doc.ents:
    print(e.text,e.start_char,e.end_char,e.label_)
```

```python
from spacy import displacy
displacy.render(doc,style="dep")
```

```python
displacy.render(doc,style="ent")
```

### POS(Parts of speech)

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
import nltk

nltk.download('punkt_tab','averaged_perceptron_tagger_eng')
```

```python
words = nltk.word_tokenize(content)
pos = pos_tag(words)
```

```python
type(pos)
```

```python
print("PoS Tagging Result:")
for word, pos_tag in pos:
    print(f"{word}: {pos_tag}")
```
