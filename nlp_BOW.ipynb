### BagOFWORDS

```python
from nltk import sent_tokenize
import re
```

```python
para = "Tesla is one of the most innovative companies in the world, " \
"known for its groundbreaking advancements in electric vehicles. " \
"Elon Musk, the CEO of Tesla, has been the driving force behind the company’s success. " \
"In addition to his role at Tesla, Musk has also significantly influenced the space industry through " \
"SpaceX, another company he founded. Tesla's mission is to accelerate the world's transition to " \
"sustainable energy, and it has garnered a dedicated team of engineers, designers, and leaders who work " \
"tirelessly to achieve this vision. " \
"Martin Eberhard, one of Tesla's co-founders, initially started the company with a vision to create " \
"high-performance electric vehicles. Although Eberhard eventually left Tesla ,his contributions are still seen as foundational to the company's growth."
```

```python
senten = sent_tokenize(para)
```

```python
for i in range(len(senten)):
    senten[i] = senten[i].lower()
    senten[i] = re.sub(r'\W','',senten[i])
    senten[i] = re.sub(r'\s+','',senten[i])

```

```python
len(senten)
```

```python
### Counting word frequency
wordcount = {}
words = word_tokenize(para)
for w in words:
    wordcount[w] = wordcount.get(w,0) + 1
    
```

```python
words
```

```python
filtered = {w : c for w,c in wordcount.items() if w  not in stop}
```

```python
type(filtered)
```

```python
senten
```

```python
da = sorted(filtered.items(),key= lambda x : x[1],reverse = True)
```

```python
da
```

```python
da = pd.DataFrame(da,columns=["words","count"])
```

```python
da
```

```python

import matplotlib.pyplot as plt
```

```python
freq = da.nlargest(10,"count")
```

```python
freq
```

```python
plt.figure(figsize=(10, 6))
plt.bar(freq["words"], freq["count"], color='skyblue')
plt.xticks(rotation=45)
plt.title('Top 10 Most Frequent Characters')
plt.xlabel('Characters')
plt.ylabel('Frequency')
```

```python
import seaborn as sns
x = []
for s in senten :
    vec = []
    for w in freq["words"]:
        if w in s:
            vec.append(1)
        else:
            vec.append(0)
    x.append(vec)


```

```python
x
```

```python
import numpy as np
x = np.asarray(x)
```

```python
x
```

```python


plt.figure(figsize=(10, 6))
sns.heatmap(x, cmap='viridis', annot=True, fmt="d", xticklabels=freq["words"], yticklabels=[f"Sentence {i+1}" for i in range(len(senten))])
plt.title('Bag of Words Matrix')
plt.xlabel('Frequent Chars')
plt.ylabel('Sentences')
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(frequencies=filtered)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation='bilinear')
#plt.title("Word Cloud of Frequent Words")
plt.axis("off")
```
